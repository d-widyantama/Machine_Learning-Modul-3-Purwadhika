```python
## EDA Standard Libary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.stats as ss
```

```python
#Loading tips dataset
tips = sns.load_dataset('tips')
```

```python
#Construct deep Info on columns & values:

NYCtaxi = []
for i in tips.columns:
    NYCtaxi.append([i, tips[i].dtypes,
                      tips[i].isna().sum(),
                      round((((tips[i].isna().sum())/(len(tips)))*100),2), 
                    tips[i].nunique(), 
                    tips[i].sample(3).values])
pd.DataFrame(NYCtaxi, columns = ['dataFeatures', 'dataType', 'null', 'nullPct', 'unique','uniqueSample'])
```

```python
#Converting categorical to dummy var

pd.get_dummies(tips, columns=['sex', 'smoker', 'time'], dtype=int, drop_first=True) #drop_first=True means eliminating the first variable and made it as basis variable
```
